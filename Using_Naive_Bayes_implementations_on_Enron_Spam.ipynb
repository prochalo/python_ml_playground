```
import glob
import os
emails = list()
labels = list()

file_path = 'enron1/spam/'
for filename in glob.glob(os.path.join(file_path, '*.txt')):
  with open(filename, 'r', encoding = "ISO-8859-1") as infile:
    emails.append(infile.read())
    labels.append(1)
```

```
file_path = 'enron1/ham/'
for filename in glob.glob(os.path.join(file_path, '*.txt')):
  with open(filename, 'r', encoding = "ISO-8859-1") as infile:
    emails.append(infile.read())
    labels.append(0)
```

```
print(len(emails))
print(len(labels))
```
